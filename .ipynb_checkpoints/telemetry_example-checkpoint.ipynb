```python
import fastai.datasets
import fastai.vision
import torch
import torch.nn as nn
import torch.nn.functional as F
from callbacks import * 
from misc import *
import core
from models import *
from functools import partial
import time
import pdb
```

```python
url = fastai.datasets.URLs.MNIST
path = fastai.datasets.untar_data(url)
data = fastai.vision.ImageDataBunch.from_folder(path, train='training', valid='testing')
```

```python
nfs = [8,16,32,32]
model = get_cnn_model(data, nfs)
loss_func = nn.CrossEntropyLoss()
```

```python
cbfs = [partial(BatchTransformXCallback, greyscale_tfm), CudaCallback, LayerActivations, GradientNorms, ProgressCallback, partial(AvgStatsCallback, accuracy)]
```

```python
learn = core.Learner(model, data, loss_func, cb_funcs=cbfs)
```

```python
learn.fit(1)
```
